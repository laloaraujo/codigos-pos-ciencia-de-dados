```r
set.seed(42)

populacao <- sample(1:100, size = 100, replace = TRUE)
tamanho <- length(populacao)
media_p <- mean(populacao)

amostra <- sample(populacao, size = 12, replace = FALSE)
media_a <- mean(amostra)

diferenca = media_p - media_a

print("População:")
print(populacao)

print("Média da população:")
print(media_p)

print("Amostra:")
print(amostra)

print("Média da amostra:")
print(media_a)

print("Diferença entre médias:")
print(diferenca)

medias <- replicate(10000, mean(sample(populacao, 12)))
hist(medias, breaks = 30)
```
