```python
import numpy as np

np.random.seed(42)

populacao = np.random.choice(range(1, 101), size=100, replace=True)

tamanho = len(populacao)

media_p = np.mean(populacao)

amostra = np.random.choice(populacao, size=12, replace=False)

media_a = round(np.mean(amostra), 2)

diferenca = round((media_p - media_a), 2)

print("População:")
print(populacao)

print(f"Média da população: {media_p}")

print("Amostra:")
print(amostra)

print(f"Média da amostra: {media_a}")

print(f"Diferença entre médias: {diferenca}")
```
