```r
set.seed(64)

#Variáveis Qualitativas Nominais
print("Variáveis Qualitativas Nominais - cor favorita")
dados1 <- c("Azul","Verde","Vermelho") #cor favorita
amostra1 <- sample(dados1, size = 8, replace = TRUE)
print(amostra1)
table(amostra1)
prop.table(table(amostra1))
moda1 <- names(which.max(table(amostra1)))
print(paste("Moda:", moda1))

#Variáveis Qualitativas Ordinais

print("Variáveis Qualitativas Ordinais - nível de satisfação")
dados2 <- c("Ruim","Regular","Bom","Excelente") #nível de satisfação
amostra2 <- sample(dados2, size = 8, replace = TRUE)
print(amostra2)
table(amostra2)
prop.table(table(amostra2))
moda2 <- names(which.max(table(amostra2)))
print(paste("Moda:", moda2))

#Variáveis Quantitativas Discretas

print("Variáveis Quantitativas Discretas - número de filhos")
dados3 <- c(0, 1, 2, 3) #número de filhos
amostra3 <- sample(dados3, size = 8, replace = TRUE)
print(amostra3)
table(amostra3)
prop.table(table(amostra3))
moda3 <- names(which.max(table(amostra3)))
print(paste("Moda:", moda3, "filhos"))

#Variáveis Quantitativas Contínuas

print("Variáveis Quantitativas Contínuas - altura")
amostra4 <- round(runif(12, min = 1.50, max = 1.80), 2)
print(amostra4)
table(amostra4)
prop.table(table(amostra4))
moda4 <- names(which.max(table(amostra4)))
print(paste("Moda:", moda4, "metros"))
media <- round(mean(amostra4), 2)
print(paste("Média:", media, "metros"))

```
