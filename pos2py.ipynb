```python
import numpy as np
from collections import Counter

# Equivalente ao set.seed(64)
np.random.seed(64)

# ==================================================
# Variáveis Qualitativas Nominais
# ==================================================

print("Variáveis Qualitativas Nominais - cor favorita")

dados1 = ["Azul", "Verde", "Vermelho"]

amostra1 = np.random.choice(dados1, size=8, replace=True)

print(amostra1)

freq1 = Counter(amostra1)

print("Frequência:")
print(freq1)

prop1 = {k: round(v / len(amostra1), 2) for k, v in freq1.items()}

print("Proporção:")
print(prop1)

moda1 = freq1.most_common(1)[0][0]

print(f"Moda: {moda1}")

# ==================================================
# Variáveis Qualitativas Ordinais
# ==================================================

print("\nVariáveis Qualitativas Ordinais - nível de satisfação")

dados2 = ["Ruim", "Regular", "Bom", "Excelente"]

amostra2 = [
    'Excelente', 'Ruim', 'Bom', 'Bom',
    'Regular', 'Ruim', 'Bom', 'Ruim'
]

freq2 = Counter(amostra2)

print(freq2)

# Maior frequência
max_freq = max(freq2.values())

# Todas as modas
modas = [k for k, v in freq2.items() if v == max_freq]

if len(modas) == 1:
    print(f"Moda: {modas[0]}")
elif len(modas) == 2:
    print(f"Bimodal: {modas}")
else:
    print(f"Multimodal: {modas}")

# ==================================================
# Variáveis Quantitativas Discretas
# ==================================================

print("\nVariáveis Quantitativas Discretas - número de filhos")

dados3 = [0, 1, 2, 3]

amostra3 = np.random.choice(dados3, size=8, replace=True)

print(amostra3)

freq3 = Counter(amostra3)

print("Frequência:")
print(freq3)

prop3 = {k: round(v / len(amostra3), 2) for k, v in freq3.items()}

print("Proporção:")
print(prop3)

moda3 = freq3.most_common(1)[0][0]

print(f"Moda: {moda3} filhos")

# ==================================================
# Variáveis Quantitativas Contínuas
# ==================================================

print("\nVariáveis Quantitativas Contínuas - altura")

amostra4 = np.round(
    np.random.uniform(1.50, 1.80, 12),
    2
)

print(amostra4)

freq4 = Counter(amostra4)

print("Frequência:")
print(freq4)

prop4 = {k: round(v / len(amostra4), 2) for k, v in freq4.items()}

print("Proporção:")
print(prop4)

moda4 = freq4.most_common(1)[0][0]

print(f"Moda: {moda4} metros")

media = round(np.mean(amostra4), 2)

print(f"Média: {media} metros")
```
